#### 1) Make a Matplotlib for below array with ```Review and TRP's``` of Broadcasting shows, by Adding Grid Lines
```
x : ([87, 81, 19, 95, 10, 105, 1100, 15, 12, 125])
y : ([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
```

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([87, 81, 19, 95, 10, 105, 1100, 15, 12, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Broadcasting of shows Data")
plt.xlabel("Reviews of shows")
plt.ylabel("TRPs of shows")

plt.plot(x, y)

plt.grid()

plt.show()
````
