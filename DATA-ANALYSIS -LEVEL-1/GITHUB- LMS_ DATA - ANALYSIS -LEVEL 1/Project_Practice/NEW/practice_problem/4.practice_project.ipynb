import pandas as pd

# Load the AirBase data into a data frame<br>
airbase_data = pd.read_csv('airbase_data.csv')

# Display the first few rows of the data frame<br>
print(airbase_data.head())

# Filter the data for a specific country<br>
country_data = airbase_data[airbase_data['Country'] == 'Germany']

# Display the filtered data<br>
print(country_data)

# Group the data by city and calculate the average air quality values<br>
city_avg_data = airbase_data.groupby('City').mean()

# Display the average values per city<br>
print(city_avg_data)

# Sort the data frame by air quality values in descending order<br>
sorted_data = airbase_data.sort_values('AirQuality', ascending=False)

# Display the sorted data<br>
print(sorted_data)

import pandas as pd

# Load the dataset<br>
data = pd.read_csv('airbase_data.csv')

# Display the first few rows of the dataset<br>
print(data.head())

# Filter the data for a specific country (e.g., France)<br>
france_data = data[data['Country'] == 'France']

# Calculate the average air quality index (AQI) for each station<br>
average_aqi = france_data.groupby('Station')['AQI'].mean()

# Sort the stations based on the average AQI in descending order<br>
sorted_stations = average_aqi.sort_values(ascending=False)

# Display the top 5 stations with the highest average AQI<br>
print(sorted_stations.head())

# Calculate the maximum and minimum AQI values in the dataset<br>
max_aqi = data['AQI'].max()<br>
min_aqi = data['AQI'].min()

print('Maximum AQI:', max_aqi)<br>
print('Minimum AQI:', min_aqi)

import pandas as pd

# Load the AirBase data into a data frame<br>
airbase_data = pd.read_csv('airbase_data.csv')

# Display the first few rows of the data frame<br>
print(airbase_data.head())

# Filter the data to include only specific countries<br>
countries = ['Germany', 'France', 'Italy']<br>
filtered_data = airbase_data[airbase_data['Country'].isin(countries)]

# Group the data by monitoring station and calculate the average air quality measurements<br>
station_avg_data = filtered_data.groupby('Station')['AirQuality'].mean()

# Sort the stations based on the average air quality measurements<br>
sorted_stations = station_avg_data.sort_values(ascending=False)

# Display the top 10 stations with the highest average air quality<br>
top_10_stations = sorted_stations.head(10)<br>
print(top_10_stations)
