### Load Dataset

```python
import pandas as pd
from sklearn.datasets import load_diabetes

# Load the diabetes dataset
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target

# Create a DataFrame using the feature data (X) and add a target column
columns = diabetes.feature_names  # Use the feature names as column names
df = pd.DataFrame(data=X, columns=columns)
df['target'] = y

# Display the DataFrame
df
```

### Train and Test Dataset (Data Modeling)

```python
x = df[['age', 'sex', 'bmi','bp','s1','s2','s3', 's4', 's5','s6']]
y = df[['target']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

### Model Training 

```python
from sklearn.linear_model import LinearRegression

lm =LinearRegression()

lm.fit(x_train,y_train)
```

```python
print(lm.intercept_)
```

```python
y_pred = lm.predict(x_test)
print(y_pred)
```

### Evaluation

```python
from sklearn.metrics import r2_score,mean_squared_error

r2 = r2_score(y_test,y_pred)

rmse = np.sqrt(mean_squared_error(y_test,y_pred))

print('R-squared', r2)
print('RMSE', rmse)
```

```python
new_data = pd.DataFrame()
new_data = (x_test)

new_data['Actual'] = y_test
new_data['Predicted'] = y_pred
```
